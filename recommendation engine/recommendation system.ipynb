```python
# converting long data into wide data using pivot table
df=data1.pivot_table(index='User_ID',columns='Book_Title',values='Book_Rating').reset_index(drop=True)
df
```

```python
# Replacing the index values by unique user Ids
df.index=data1['User_ID'].unique()
df
```

```python
# Impute those NaNs with 0 values
df.fillna(0,inplace=True)
df
```

```python
# Calculating Cosine Similarity between Users on array data
user_sim=1-pairwise_distances(df.values,metric='cosine')
user_sim
```

```python
# Store the results in a dataframe format
user_sim2=pd.DataFrame(user_sim)
user_sim2
```

```python
# Set the index and column names to user ids 
user_sim2.index=data1['User_ID'].unique()
user_sim2.columns=data1['User_ID'].unique()
user_sim2
```

```python
# Nullifying diagonal values
np.fill_diagonal(user_sim,0)
user_sim2
```

```python
# Most Similar Users
user_sim2.idxmax(axis=1)
```
